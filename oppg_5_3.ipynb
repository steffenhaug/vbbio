# Enkelt potensial
Målet er å simulere diffusjon i eit enkelt potensial

$$ V(x) = kx $$

for ein konstant $k$.

```python
import scipy as scp
import scipy.constants as const
k_B = const.Boltzmann
```

Det er oppgitt at $k\beta$ er mellom $3$ og $5$.

```python
temp = 300
beta = 1 / (k_B * temp)
k = 3 / beta

# step length
h = 0.5

def V(x):
    if x < -3*h:
        return -k
    elif x > 3 * h:
        return k
    else:
        return k * (-1 + 2 * (x + 3*h) / (6*h))

def P_r(xs):
    rs = np.zeros(xs.shape)
    for i, x in enumerate(xs):
        rs[i] = np.exp(-beta * (V(x + h) - V(x - h)))
    return 1 / (1 + rs)

def steps(xs):
    rng = np.random.random(xs.shape)
    return h * np.where(rng > P_r(xs), 1, -1)
```

Gitt $x$-verdiane kan vi beregne alle $P_R(x)$-verdiane.
Desse verdiane brukar vi for å ta steg tilfeldig.

```python
# start parameters: N = #particles, T = #steps
N = 10000
T = 500

plot_ts = [5, 50, 150, 500]

# N partiklar i starttilstand
pos = np.empty((N,))
pos[::2]  = +10 * h
pos[1::2] = -10 * h

phi = dict()

for t in range(T + 1):
    pos += steps(pos)
    if t in plot_ts:
        phi[t] = np.copy(pos)
```

```python
fig = plt.figure()
axes = fig.subplots(1 + len(plot_ts), 1, sharex=True)

ylim = (0, .5)

### plot 1
for i, t in enumerate(plot_ts):
    unique, count = np.unique(phi[t], return_counts=True)
    axes[i].bar(
        unique, count / N,
        edgecolor="black",
        color='gray',
        width=2*h,
        align='center',
        label='Empirisk. $t = {}$'.format(t),
    )

    axes[i].set_ylim(*ylim)
    axes[i].set_ylabel(r"Sannsyn")  
    axes[i].legend(loc='upper right')


### plot 2
xmin, xmax = plt.xlim()

xs = np.linspace(xmin, xmax, 500)
vs = np.zeros(xs.shape)
for i, x in enumerate(xs):
    vs[i] = V(x)

axes[-1].plot(xs, vs, 'k', label="$V(x)$")
axes[-1].set_ylabel(r"Potensial")
axes[-1].legend(loc='lower right')
axes[-1].set_ylim(1.1 * np.min(vs), np.max(vs) * 1.1)

plt.xlabel(r"Posisjon")

plt.savefig("walk_5_3.pdf")
```
